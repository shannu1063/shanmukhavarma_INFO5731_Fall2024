## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import pos_tag, word_tokenize
from sklearn.feature_extraction.text import TfidfTransformer
import nltk

# Few sample data
texts = [
    "The movie was fantastic and I loved it!",
    "It was a terrible film, really disappointing.",
    "Not good, I wouldn't recommend it to anyone."
]

#Feature Extraction
# 1. Bag of Words (BoW)
vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(texts)
print("Bag of Words Feature Matrix:\n", X_bow.toarray())
print("BoW Feature Names:\n", vectorizer.get_feature_names_out())

# 2. TF-IDF
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(texts)
print("TF-IDF Feature Matrix:\n", X_tfidf.toarray())
print("TF-IDF Feature Names:\n", tfidf_vectorizer.get_feature_names_out())

# 3. Sentiment Lexicon Scores using VADER
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()
for text in texts:
    sentiment_score = sid.polarity_scores(text)
    print(f"Sentiment score for '{text}':", sentiment_score)

# 4. Tagging of Part of Speech (POS)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
for text in texts:
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    print(f"POS Tags for '{text}':", pos_tags)

# 5. N-grams (Bigrams)
vectorizer_ngrams = CountVectorizer(ngram_range=(2, 2))
X_bigrams = vectorizer_ngrams.fit_transform(texts)
print("Bigrams Feature Matrix:\n", X_bigrams.toarray())
print("Bigrams Feature Names:\n", vectorizer_ngrams.get_feature_names_out())

```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
# You code here (Please add comments in the code):

from sklearn.feature_selection import chi2
import numpy as np

# Creating labels (1: positive, 0: negative)
y = [1, 0, 0]

# Feature selection based on Chi-square  on TF-IDF matrix
chi2_scores, p_values = chi2(X_tfidf, y)

# Ranking features (highest to lowest chi-square score)
ranking_indices = np.argsort(chi2_scores)[::-1]
ranked_features = np.array(tfidf_vectorizer.get_feature_names_out())[ranking_indices]

print("Ranked Features based on Chi-Square Scores:\n", ranked_features)
print("Chi-Square Scores:\n", chi2_scores[ranking_indices])


```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
# You code here (Please add comments in the code):

from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Loading the already pre-trained BERT model and also tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_bert_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze()

# An example of Sample query
query = "The movie was amazing!"

# Extracting BERT embeddings for the queries and texts involved
text_embeddings = [get_bert_embeddings(text) for text in texts]
query_embedding = get_bert_embeddings(query)

# Calculating cosine similarity
cosine_similarities = cosine_similarity(
    [query_embedding],
    [embedding.numpy() for embedding in text_embeddings]
)

# Ranking the texts based on similarity
ranked_indices = np.argsort(cosine_similarities[0])[::-1]
print("Ranked Texts by Similarity:")
for index in ranked_indices:
    print(f"Text: {texts[index]}, Similarity Score: {cosine_similarities[0][index]}")


```
